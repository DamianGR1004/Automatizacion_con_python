# Día 8: Visualización básica con matplotlib

### Objetivo
Aprender a crear gráficos sencillos (lineas y barras) para represntar datos de forma visual

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vamos a crear un dataset de muestra

productos = ['Laptop', 'Mouse', 'Monitor', 'Teclado', 'Audifonos']
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']

df = pd.DataFrame({
    'prodcuto': np.random.choice(productos, 100),
    'mes': np.random.choice(meses, 100),
    'ventas': np.random.randint(100, 1000, 100)
})
```

```python
df.head()
```

#### Top 5 productos vendidos

```python
top_5 = df.groupby('prodcuto')['ventas'].sum().sort_values(ascending=False).head(5)
```

#### Gráfico de barras

```python
plt.figure(figsize=(10, 6))
barras = plt.bar(top_5.index, top_5.values, color='#0b466c')
plt.title('Top 5 productos vendidos')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.grid(True)

# Para agregar eqiquetas de datos

for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 30,  # posición
             f'{altura}', ha='center', va='bottom', fontsize=9)

plt.show()
```

#### Gráfico de líneas

```python
# Vamos a identificar las ventas por mes

ventas_por_mes = df.groupby('mes')['ventas'].sum()
```

```python
plt.figure(figsize=(10, 6))
lineas = plt.plot(ventas_por_mes, marker='o', linestyle='-', color='#0b466c')
plt.title('Ventas por mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid(True)

# Ver los valores en cada punto

for i, valor in enumerate(ventas_por_mes):
    plt.text(ventas_por_mes.index[i], valor + 100, f'{valor}', ha='center', va='bottom', fontsize=9)

plt.show()
```
